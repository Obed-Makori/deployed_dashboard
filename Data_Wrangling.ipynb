```python
# import the required libraries for data wrangling

#data manipulation modules
import numpy as np
import pandas as pd

# data visualization modules
import matplotlib.pyplot as plt
import seaborn as sns

# warnings ignorance
import warnings
warnings.filterwarnings("ignore")
```

```python
#initialize url for the data
url="https://data.kcmo.org/api/views/vsgj-uufz/rows.csv?accessType=DOWNLOAD"
```

```python
# pass the data in the url into a pandas dataframe
data=pd.read_csv(url)
# display the first 10 rows of the data, to display the last 10 use data.tail()
data.head(10)
```

```python
# check the shape of the data
data.shape
print("the data contains ", data.shape[0], "rows and",data.shape[1],"columns")
```

```python
# missing values
data.isna().sum()
```

```python
#data summary
data.describe
```

```python
# statistical summary
data.describe()
```

```python
# datatypes
data.dtypes
```

```python
# cleaning

# drop missing values 
data_1=data.dropna()
# data_ is the new clean dataset. to mantain we use the parametre inplace=true
data_1.head(5)
```

## Similar data

```python
# to combine with the similar dataset , we may use concatenate or merge python functionns to combine the two datasets
url2='https://data.kcmo.org/api/views/pxaa-ahcm/rows.csv?accessType=DOWNLOAD'
# save the data in the url into a pandas dataframe
data_2=pd.read_csv(url2)
# display 5 samples of rows in the dataset
data_2.sample(5)
```

```python
# CLEANING THE SECOND DATA

#Drop missing values
data_2.dropna(inplace=True)
# display the clean data
data_2.head(5)
```

```python
# understanding data2
data_2.describe
```

```python
# statistical summary of data2
data_2.describe()
```

```python
# data2 datatypes of columns
data_2.dtypes
```

```python
# the two dataset can now be combined in order to provide detailed analysis

all_data=pd.concat([data_1, data_2], ignore_index=True)
```

```python
# displaying the first 5 rows of the combined data
all_data.head(5)
```

```python
# summary of the combined data
all_data.describe
```
